```python
import altair as alt
import pandas as pd
import matplotlib.pyplot as mplt
import geopandas as gp
import numpy as np
from shapely.geometry import Point
import missingno as msn
import seaborn as sns

bike_data = pd.read_csv('BikeData.csv')

bike_data.head()
```

```python
latest_tfl_bikes = pd.read_csv('LatestTFLBikes - 2020-10-07 to 2020-26-01.csv')

latest_tfl_bikes = latest_tfl_bikes.rename(columns=({'id':'stationId'}))

latest_tfl_bikes.head()
```

```python
bikedata_by_tfldata_bikeid = pd.merge(bike_data, latest_tfl_bikes, on='stationId')

bikedata_by_tfldata_bikeid
```

```python
# What can we derive from our summary statistics above?

# If we look at 'availableBikes', the average is 11 and it has a std deviation of around 9.
# You could argue that there is a high std deviation for 'availableBikes' since it is almost has high as the average.
# The same goes for 'availableDocks' and 'numBikes' and 'numEmptyDocks'.
# All the variables with the exception of the 'lat' and 'long' have high ranges which suggests the data has low central tendency.

# Why don't have a look at the distribution to check our assumptions?
histogram_available_bikes = mplt.hist(bikedata_by_tfldata_bikeid.availableBikes, bins=10)
mplt.title('Available Bikes Histogram')
mplt.xlabel('Available Bikes')
mplt.ylabel('Count')

# We can clearly see that 'availableBikes' do not have central tendency. 
# The values are heavily skewed to the left, i.e. positive skew with most available bikes being around 0 to 10.
```

```python
histogram_available_docks = mplt.hist(bikedata_by_tfldata_bikeid.availableDocks, bins=10)

# Docks are also skewed to the left but the average number of docks available seems to be around 15.
```

```python
# The below chart does not really tell us much!
histogram_available_bikes_vs_docks = mplt.scatter(bikedata_by_tfldata_bikeid.stationId, bikedata_by_tfldata_bikeid.availableDocks)
```

```python
# We need to use the 'lat' (latitude) and 'lon' (longitude) columns to position our data in a shape file.
# Latitude is a geographic coordinate that specifies the north–south position of a point on the Earth's surface. 
# Latitude is an angle which ranges from 0° at the Equator to 90° (positive for north and negative for south) at the poles.

# Longitude, is a geographic coordinate that specifies the east–west position of a point on the Earth's surface.
# The half point is 180° with 0 to 179° being the east and 0 to -179° being the west side.
```

```python
bikedata_by_tfldata_bikeid['coordinates'] = bikedata_by_tfldata_bikeid[['long', 'lat']].values.tolist()

bikedata_by_tfldata_bikeid['coordinates'] = bikedata_by_tfldata_bikeid['coordinates'].apply(Point)

bikedata_by_tfldata_bikeid.head()
```

```python
bikedata_by_tfldata_bikeid['coordinates']
```

```python
# Let us now create our 'choropleth map'.
# A choropleth map is  type of thematic map in which areas are shaded 
# or patterned in proportion to a statistical variable that represents 
# an aggregate summary of a geographic characteristic within each area, 
# such as population density or per-capita income.

# First join the geodataframe with the csv dataframe
# 'set_index' sets the index value using the values of one or more existing columns
# Good resource for understanding the method can be found here:
# https://www.w3resource.com/pandas/dataframe/dataframe-set_index.php#:~:text=The%20set_index()%20function%20is,index%20or%20expand%20on%20it.

bikedata_by_tfldata_bikeid = gp.GeoDataFrame(bikedata_by_tfldata_bikeid, geometry='coordinates')

bikedata_by_tfldata_bikeid.plot(figsize=(20, 10))
```

```python
# Finally, we create a figure and axes for it using Matplotlib

fig, ax = mplt.subplots(1, figsize=(30, 20)) # The 'subplots' method helps us to create subplots, i.e. plots within a plot

map_df = map_df.to_crs(epsg=4326) # IMPORTANT! Convert London Map Coordinates To Lat and Long

base = map_df.plot(ax=ax, color='#3B3C6E')

bikedata_by_tfldata_bikeid.plot(ax=base, color='darkred', linewidth=0.8, edgecolor='0.8', marker="*", markersize=10)
```

```python
# Let us check if the above plot gives accurate results.


availableBikes_by_area_name = bikedata_by_tfldata_bikeid.groupby(bikedata_by_tfldata_bikeid['name']).aggregate({'availableBikes': 'sum'})

availableBikes_by_area_name.head()
```
